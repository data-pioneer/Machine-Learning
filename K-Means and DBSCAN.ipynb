```python
# import necessary libraries 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN 
```

##### ●  Read about K-Means and DBSCAN algorithms and understand their working principles and differences.

K-means and DBSCAN are both used to group datasets into specific number of groups which are called clusters. DBSCAN stands for Density-Based Spatial Clustering algorithm.It handles outliers in a better way with irregularly shaped clusters. This application does not apply in the case of K-means which works efficiently when the clusters are spherical, equally sized, and have a similar density.

##### ●  Generate a synthetic dataset with 3-4 clusters using Scikit-Learn's make_blobs function.

```python
# Create blob Data
X, y  = make_blobs(n_samples=2000, 
                           centers=4, cluster_std=1.8,random_state=42)
```

```python
# create scatter plot to display blob data in the form of clusters
plt.scatter(X[:, 0], X[:, 1],c= X[:, 1] , cmap='rainbow') 
```

##### ●  Implement K-Means clustering algorithm on the generated dataset with varying values of K, i.e., 2, 3, 4, and visualize the resulting clusters using Matplotlib.

```python
# implementation of Kmeans with 2 Clusters
kmeans_C2 = KMeans(n_clusters=2)
kmeans_C2.fit(X)
```

```python
# Display scatter plot of k means and original custer for more clarity
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(X[:,0],X[:,1],c=kmeans_C2.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(X[:,0],X[:,1],c=X[:,1],cmap='rainbow')
```

```python
# Kmeans with 3 Clusters
kmeans_C3 = KMeans(n_clusters=3)
kmeans_C3.fit(X)
```

```python
# Display scatter plot of k means and original custer for more clarity
f, (ax1, ax2) = plt.subplots(1, 2,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(X[:,0],X[:,1],c=kmeans_C3.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(X[:,0],X[:,1],c=X[:,1],cmap='rainbow')
```

```python
# Kmeans with 4 Clusters
kmeans_C4 = KMeans(n_clusters=4)
kmeans_C4.fit(X)
```

```python
# Display scatter plot of k means and original custer for more clarity
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(X[:,0],X[:,1],c=kmeans_C4.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(X[:,0],X[:,1],c=X[:,1],cmap='rainbow')
```

##### ●  Evaluate the K-Means model using Silhouette score and determine the optimal number of clusters.

```python
kmeans_C2_Silhousette_score= silhouette_score(X, kmeans_C2.labels_)
print("k2 : ",kmeans_C2_Silhousette_score)
```

```python
kmeans_C3_Silhousette_score= silhouette_score(X, kmeans_C3.labels_)
print("k3 : ",kmeans_C3_Silhousette_score)
```

```python
kmeans_C4_Silhousette_score= silhouette_score(X, kmeans_C4.labels_)
print("k4 : ",kmeans_C4_Silhousette_score)
```

### ●   Implement DBSCAN clustering algorithm on the same dataset and visualize the resulting clusters.

```python
epsilon = 0.3
minimumSamples = 7
db_Min_Sample7 = DBSCAN(eps=epsilon, min_samples=minimumSamples).fit(X)
```

```python
plt.scatter(X[:,0],X[:,1], c = db_Min_Sample7.labels_)
```

```python
db_Min_Sample7_SScore = silhouette_score(X, db_Min_Sample7.labels_)
print("epsilon = 0.3,\tminimumSamples = 7, silhouette_score =",db_Min_Sample7_SScore)
```

##### ●  Experiment with different values of hyperparameters, i.e., eps and min_samples, and observe their effect on the resulting clusters. 

```python
epsilon = 0.7
minimumSamples = 30
db_Min_Sample30 = DBSCAN(eps=epsilon, min_samples=minimumSamples).fit(X)
```

```python
db_Min_Sample30_E7_SScore =  silhouette_score(X, db_Min_Sample30.labels_)
print("epsilon = 0.7,\tminimumSamples = 30, silhouette_score =",db_Min_Sample30_E7_SScore)
```

```python
plt.scatter(X[:,0],X[:,1], c = db_Min_Sample30.labels_)
```

```python
epsilon = 0.9
minimumSamples = 30
db_Min_Sample30 = DBSCAN(eps=epsilon, min_samples=minimumSamples).fit(X)
```

```python
db_Min_Sample30_E9_SScore =  silhouette_score(X, db_Min_Sample30.labels_)
print("epsilon = 0.9,\t minimumSamples = 30,\tSilhouette_score =",db_Min_Sample30_E9_SScore)
```

```python
plt.scatter(X[:,0],X[:,1], c = db_Min_Sample30.labels_)
```
